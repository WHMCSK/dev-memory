```python
import numpy as np
import wave
from IPython.display import Audio
import matplotlib.pyplot as plt
import librosa

# 参数设置
sample_rate = 44100  # 采样率，每秒样本数
duration = 15  # 音频持续时间（秒）
freq = 440  # 正弦波的频率（赫兹）

# 生成时间轴
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# 生成正弦波信号
audio_signal = (0.5 * np.sin(2 * np.pi * freq * t)).astype(np.float32)

plt.figure(1)
plt.plot(audio_signal)
plt.title('Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.savefig('音频频率图.png')
plt.show()

# 将信号转换为16位整数
audio_signal = np.int16(audio_signal * 32767)

# 创建WAV文件
wav_file = wave.open('sine_wave.wav', 'wb')
wav_file.setnchannels(1)  # 单声道
wav_file.setsampwidth(2)  # 2 bytes per sample
wav_file.setframerate(sample_rate)
wav_file.writeframes(audio_signal.tobytes())
wav_file.close()

print('音频文件已保存为 sine_wave.wav')

y, sr = librosa.load('sine_wave.wav')

plt.figure(2)
plt.plot(y)
plt.title('Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()

# 在Jupyter Notebook中播放声音
Audio(data=audio_signal, rate=sample_rate)
```
